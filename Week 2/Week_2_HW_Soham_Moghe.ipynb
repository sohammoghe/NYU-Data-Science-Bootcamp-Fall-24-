## Practice Questions for numpy

1. Define two custom numpy arrays, say A and B. Generate two new numpy arrays by stacking A and B vertically and horizontally.

```python
import numpy as np
```

```python
A = np.arange(5)
print(f"""
A:
{A}
Shape: {A.shape}
Type: {type(A)}
""")

B = np.arange(2,7)
print(f"""
B:
{B}
Shape: {B.shape}
Type: {type(B)}
""")

new_vstack = np.vstack((A,B))
print(f"""
Vertically Stacked (A and B):
{new_vstack}
Shape: {new_vstack.shape}
Type: {type(new_vstack)}
""")

new_hstack = np.hstack((A,B))
print(f"""
Horizontally Stacked (A and B):
{new_hstack}
Shape: {new_hstack.shape}
Type: {type(new_hstack)}
""")
```

2. Find common elements between A and B. [Hint : Intersection of two sets]

```python
common_elements, common_indices_A,  common_indices_B  = np.intersect1d(A, B, return_indices = True)
# interesct1d finds common elements between A and B by treating them as sets
# It can also return the indices of common elements

print(f"""Common elements between A and B: {common_elements}
Type of common elements generated: {type(common_elements)}
Indices of common elements in A: {common_indices_A} Type: {type(common_indices_A)}
Indices of common elements in B: {common_indices_B} Type: {type(common_indices_B)}
""")
```

3. Extract all numbers from A which are within a specific range. eg between 5 and 10. [Hint: np.where() might be useful or boolean masks]

```python
#Extracting numbers from A between 2 and 8 (inclusive of both) using np.where()
low = 2
high = 8
extraction_result =  A[np.where((A >= low) & (A <= high))]
print(f"Numbers in A from 2 to 8 (inclusive of both): {extraction_result}")
```

4. Filter the rows of iris_2d that has petallength (3rd column) > 1.5 and sepallength (1st column) < 5.0

```python
url = 'https://archive.ics.uci.edu/ml/machine-learning-databases/iris/iris.data'
iris_2d = np.genfromtxt(url, delimiter=',', dtype='float', usecols=[0,1,2,3])
iris_2d
```

```python
print(type(iris_2d))
print(iris_2d.shape)
```

```python
# Extracting Petal Length Column values (3rd Column)
petal_length = iris_2d[:, 2]
petal_length
```

```python
# Extracting Sepal Length Column values (1st Column)
sepal_length = iris_2d[:, 0]
sepal_length
```

```python
#Samples having petallength (3rd column) > 1.5 and sepallength (1st column) < 5.0
filtered_samples = iris_2d[np.where((petal_length > 1.5) & (sepal_length < 5.0))]
filtered_samples
```
